```python
from transformers import pipeline

# class1 = pipeline('text-classification', model='nlptown/bert-base-multilingual-uncased-sentiment')
class1 = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')
result1 = class1("The level of alignment between my expectations and the closing two installments of the series remained somewhat indeterminate.")
result2 = class1("The culmination offered by the series’ final pair of chapters produced an experiential contour that diverged from the anticipated trajectory.")
result3 = class1("Within the terminal episodes, the interplay between narrative momentum and viewer expectation manifested in a manner less convergent than presumed.")
result4 = class1("The resonance elicited by the concluding installments seemed suspended in a space where anticipation and realization did not fully converge.")
print(result1,result2,result3,result4,sep = "\n\n")
########3 stars (Neutral)
# 4 stars (Positive)
# 3 stars (Neutral)
# 2 stars (Negative)
```

```python
list_sentence = ["The humor in TMKOC always lightens my mood and makes the day feel brighter.",#+ve
                 "Sometimes the storylines feel dragged out and lose the charm they once had.",#-ve
                 "TMKOC has been running for years, and it continues to hold a special place in Indian television.",#neutral
                 "The chemistry between Jethalal and Daya is unmatched and always entertaining.",#+ve
                 "Despite its popularity, certain episodes seem repetitive and less engaging."#negative
                 ]
sa = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")
sa(list_sentence)
```

```python
text_generator = pipeline("text-generation", model="distilgpt2")

outputs = text_generator(
    "Today is rainy day in Indore so",
    num_return_sequences=3,
    truncation = True,
    do_sample=True,      # enables randomness
    temperature=0.9,     # lower = more focused, higher = more creative
    top_k=50,            # only sample from top 50 candidates
    top_p=0.95           # nucleus sampling (cumulative probability)
)

for i, out in enumerate(outputs, 1):
    print(f"Output {i}: {out['generated_text']}\n")
```

```python
qa_model = pipeline("question-answering")

# context = "The first law of thermodynamics says energy cannot be created or destroyed."
context = """In physics, the first law of thermodynamics is also called the
law of conservation of energy. It states that energy cannot be created or destroyed,
only transformed from one form to another."""


question = "Which law of thermodynamics explains that energy cannot be created or destroyed?"

answer = qa_model(question=question, context=context)
print(answer)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
s_data = pd.read_csv('/content/drive/MyDrive/Untitled folder/Copy of sonar data.csv',header = None)
```

```python
s_data.head()
```

```python
s_data.shape
```

```python
s_data.describe()
```

```python
s_data[60].value_counts()
```

```python
s_data.groupby(60).mean()#mean value of all the columns group by rock and mine
```

```python
#seperating data and labels
x = s_data.drop(columns=60,axis = 1)
y = s_data[60]
print(x)
print(y)
```

```python
#split into training and test data
x_train , x_test , y_train,y_test = train_test_split(x,y,test_size = 0.14,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#we will use logistic regression model in training
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
#we will test the model aginst training data itself and find the accuracy then test it against test data
#accuracy >70 % is good
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print(training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
print(test_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
print(test_data_accuracy)
```

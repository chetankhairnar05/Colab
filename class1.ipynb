Handeling Outliers: extereme values are called outliers: values beyond range

IQR = Q3-Q1

OUTLIER = (-INFINITY, Q1-1.5*IQR) UNION (Q3+1.5*IQR,INFINITY)





```python
sns.boxplot(x = 'Present_Price',data = df)
```

```python
print(df['Present_Price'].quantile(0.05))
print(df['Present_Price'].quantile(0.95))
```

```python
df['Present_Price'] = np.where(df['Present_Price'] > 22.7,22.7,df['Present_Price'])
df['Present_Price'] = np.where(df['Present_Price'] < 0.54,0.54,df['Present_Price'])
```

```python
sns.boxplot(x = 'Present_Price',data = df)
```

```python
numerical_cols = df.select_dtypes(include=['int64','float64','number']).columns
features_with_outliers = []
for col in numerical_cols:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3-Q1
  lower_bound = Q1 - 1.5*IQR
  upper_bound = Q3 + 1.5*IQR

  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  if not outliers.empty:
    features_with_outliers.append(col)

print("features_with_outliers")
print(features_with_outliers)


```

```python
for cols in features_with_outliers:
  L = df[cols].quantile(0.05)
  U = df[cols].quantile(0.95)
  df[cols] = df[cols].clip(lower=L,upper=U)
```

```python
for cols in features_with_outliers:
  plt.figure(figsize=(6,4))
  sns.boxplot(x = cols,data = df)
  plt.legend()
  plt.title(f"Box_Plot of {cols}")
  plt.show()
```

In machine learning (specifically Scikit-Learn),  computes preprocessing parameters (e.g., mean, standard deviation) from training data,  applies these parameters, and  combines both for efficiency. Data leakage occurs when test data information inadvertently influences the training process, causing over-optimistic performance, similar to seeing exam questions beforehand. [1, 2, 3, 4, 5]  
Key Definitions and Use Cases

 fit(): Computes the necessary parameters (e.g., min, max, mean, standard deviation) for scaling or imputation on a dataset.

	 Use Case: Calculating the mean of a training set to be used later for scaling.

 fit_transform(): Combines  and  into one step. It calculates the parameters and then applies them to the input data.

	 Use Case: Used on the training set during preprocessing (e.g., ).

 transform(): Applies precomputed parameters (from a previous ) to new data.

	 Use Case: Used on the test/validation set to ensure it is scaled using the training datas properties (). [1, 2, 3, 6, 7]  

Data Leakage (Information Leakage)Data leakage happens when information from outside the training dataset (such as validation or test data) is used to create the model. This leads to high accuracy during training/testing but poor performance on real-world, unseen data.

 Causes:

	 Applying  on the entire dataset before splitting into train/test sets.
	 Using target information (labels) in input features.

 Preventive Measures:

	 Always split data into train/test sets before any pre-processing.
	 Use  only on the training data.
	 Use only  on the test data. [2, 5, 7, 10]  

Summary Table

| Method [2, 7, 11] | Purpose | Where to Use  |
| --- | --- | --- |
|  | Learn parameters (e.g., $\mu, \sigma$) | Training Set  |
|  | Apply parameters | Test/Validation Set  |
|  | Learn + Apply parameters | Training Set  |

Properly using  on train and  on test ensures the model generalizes well and avoids data leakage. [5, 7]

```python
import os
import matplotlib.pyplot as plt
import cv2

# Path to folder A
base_path = "/content/bovine_dataset"

# List subfolders inside A (should contain B)
folders_A = sorted(os.listdir(base_path))
folder_B = os.path.join(base_path, folders_A[0])   # pick first folder B

# List subfolders inside B (breeds)
breed_folders = sorted(os.listdir(folder_B))
first_breed = os.path.join(folder_B, breed_folders[0])  # pick first breed folder

# List images inside the first breed folder
images = sorted(os.listdir(first_breed))

# Show first 5 images
plt.figure(figsize=(15,5))
for i, img_name in enumerate(images[:5]):
    img_path = os.path.join(first_breed, img_name)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(1,5,i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Image {i+1}")
plt.show()
```

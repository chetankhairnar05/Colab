```python
# prompt: import px , plt , sns

import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Global settings
plt.rcParams['figure.figsize'] = (8, 8)  # default figure size
plt.rcParams['axes.titlesize'] = 16      # title font size
plt.rcParams['axes.labelsize'] = 14      # label font size
plt.rcParams['xtick.labelsize'] = 12     # x-tick font size
plt.rcParams['ytick.labelsize'] = 12     # y-tick font size
plt.rcParams['legend.fontsize'] = 12     # legend font size
plt.rcParams['figure.dpi'] = 100         # resolution


sns.set_theme(style='whitegrid')   # style: 'whitegrid', 'darkgrid', 'white', 'dark', 'ticks'
sns.set_context('notebook', font_scale=1.2)  # 'paper', 'notebook', 'talk', 'poster'

```

```python
fig  = px.histogram(df, x = 'age',marginal='box',nbins = 47 , title='Age Distribution')
fig.update_layout(bargap = 0.1)
fig.show()
```

```python
fig  = px.histogram(df, x = 'bmi',marginal='box',nbins = 47 , title='Age Distribution')
fig.update_layout(bargap = 0.1)
fig.show()
```

```python
fig  = px.histogram(df, x = 'charges',marginal='box',color = 'smoker',color_discrete_sequence = ['green','grey'] , title='Annual medical charges')
fig.update_layout(bargap = 0.1)
fig.show()
```

```python
fig  = px.histogram(df, x = 'smoker',marginal='box',color = 'sex',color_discrete_sequence = ['green','grey'] , title='sex vs smoker')
fig.update_layout(bargap = 0.1)
fig.show()
```

```python
fig = px.scatter_matrix(df)
fig.show()
```

```python
df['smoker_1'] = df['smoker'].apply(lambda x: 1 if x == 'yes' else 0)
df['sex_1'] = df['sex'].apply(lambda x: 1 if x == 'male' else 0)
df['region_1'] = df['region'].apply(lambda x: 0 if x == 'southwest' else 1 if x == 'southeast' else 2 if x == 'northwest' else 4)
```

```python
df.head()
```

```python
fig = px.scatter(
    df,
    x='bmi',           # X-axis
    y='charges',           # Y-axis
    color='region_1',       # Color encoding
    size='smoker_1',        # Size encoding
    symbol='sex_1',      # Symbol encoding (like x, circle, etc.)
    title="Multivariate Scatter Plot"
)
fig.show()
```

```python
fig = px.scatter(df , x = 'age',y = 'charges',color = 'smoker',opacity = 0.5,hover_data=['sex','bmi','children'])
fig.update_traces(marker_size=4)
fig.show()
```
